```python
D_sanpham = {"SP01":"Cơm gà","SP02":"Cơm sườn"}

def hienthisanpham(D_sanpham):
    ### Hiển thị sản phẩm theo dạng như sau:
    ### ---------------------------------------
    ###                                         
    ### SP01: Cơm gà
    ### SP02: Cơm sườn
    ###         
    ### ---------------------------------------
    print("---------------------------------------")
    print("                                       ")
    for i in D_sanpham:
        print(i ,":", D_sanpham[i]) 
    print("                                       ")
    print("---------------------------------------") 
    
def kt_format(id_sp):
    if len(id_sp) != 4:
        return False
    if id_sp[0:2] != "SP":
        return False
    if not id_sp[2:].isnumeric():
        return False
    return True

def themsanpham(D_sanpham):
    ### Yêu cầu người dùng nhập id sản phẩm
    ### Người dùng phải nhập đúng dạng SPXX, nếu nhập sai yêu cầu nhập lại
    ### Nếu người dùng nhập đúng dạng, thì kiểm tra xem ID đã tồn tại trong danh sách sản phẩm
    ### Nếu như ID đã tồn tại, yêu cầu người dùng nhập lại
    ### Trong quá trình nhập, nếu người dùng muốn thoát chức năng thêm sản phẩm thì nhập 0
    ### Nếu người dùng nhập ID chưa tồn tại thì yêu cầu người dùng nhập tên sản phẩm
    ### Sau khi người dùng nhập tên sản phẩm thì thêm sản phẩm vào danh sách sản phẩm
    
    id_sp = input("Nhap id cua san pham: ")
    
    while not kt_format(id_sp) or id_sp in D_sanpham:
        print("Ban vui long nhap dung cu phap va chac chan ID nay chua ton tai trong danh sach")
        if id_sp == "0":
            return
        id_sp = input("Nhap id cua san pham: ")
    name = input("Ten san pham: ")
    D_sanpham[id_sp] = name

def suatensanpham(D_sanpham):
    ### Yêu cầu người dùng nhập id sản phẩm
    ### Kiểm tra ID sản phẩm có tồn tại hay không, nếu không tồn tại, yêu cầu người nhập lại
    ### Trong quá trình nhập, nếu người dùng muốn thoát chức năng sửa sản phẩm thì nhập 0
    ### Nếu ID có tồn tại, yêu cầu người dùng nhập tên mới muốn thay thế
    ### Sau khi nhập tên mới, thì sửa tên mới trong danh sách sản phẩm
    id_sp = input("Nhap id cua san pham: ")
    
    while not id_sp in D_sanpham:
        if id_sp == "0":
            return
        print("Vui long nhap dung id san pham")
        id_sp = input("Nhap id cua san pham: ")
    

    name = input("Ten san pham: ")
    D_sanpham[id_sp] = name
    
    
def xoasanpham(D_sanpham):
    ### Yêu cầu người dùng nhập id sản phẩm
    ### Kiểm tra ID sản phẩm có tồn tại hay không, nếu không tồn tại, yêu cầu người nhập lại
    ### Trong quá trình nhập, nếu người dùng muốn thoát chức năng sửa sản phẩm thì nhập 0
    ### Nếu ID có tồn tại thì xóa sản phẩm
    id_sp = input("Nhap id cua san pham: ")
    
    while not id_sp in D_sanpham:
        if id_sp == "0":
            return
        print("Vui long nhap dung id san pham")
        id_sp = input("Nhap id cua san pham: ")
        

    del(D_sanpham[id_sp])

while True:
    print("Nhập H nếu bạn muốn hiển thị danh sách sản phẩm")
    print("Nhập T nếu bạn muốn thêm sản phẩm")
    print("Nhập S nếu bạn muốn sửa sản phẩm")
    print("Nhập X nếu bạn muốn xóa sản phẩm")
    print("Nhập 0 nếu bạn muốn dừng chương trình")
    command = input("Lệnh của bạn:")
    if command == "H":
        hienthisanpham(D_sanpham)
    elif command == "T":
        themsanpham(D_sanpham)
    elif command == "S":
        suatensanpham(D_sanpham)
    elif command == "X":
        xoasanpham(D_sanpham)
    elif command == "0":
        break

```
