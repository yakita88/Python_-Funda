```python
def rectangular(a):
    for i in range(2):
        t.fd(a)
        t.lt(90)
        t.fd(1.5*a)
        t.lt(90)
def square(a):
    for i in range(4):
        t.fd(a)
        t.lt(90)




import turtle
t = turtle.Turtle()
t.pensize(5)
a = int(input("Nhap a: "))
t.penup()
t.goto(-100,-100)
t.pendown()

t.color("blue","red")
t.begin_fill()
rectangular(a)
t.end_fill()

t.color("yellow","white")
t.begin_fill()
t.penup()
t.goto((-100+0.3*a),-1/2*a)
t.pendown()
rectangular(a/3)
t.end_fill()

t.penup()
t.goto((-100+a),-100)
t.pendown()
t.color("blue","red")
t.begin_fill()
rectangular(a)
t.end_fill()

t.color("yellow","white")
t.begin_fill()
t.penup()
t.goto((-100+1.3*a),-1/2*a)
t.pendown()
rectangular(a/3)
t.end_fill()

t.penup()
t.goto((-100),-100 + 1.5*a)
t.pendown()
t.color("blue","red")
t.begin_fill()
rectangular(a)
t.end_fill()

t.color("yellow","white")
t.begin_fill()
t.penup()
t.goto((-100+0.3*a),a)
t.pendown()
rectangular(a/3)
t.end_fill()

t.penup()
t.goto((-100+a),-100 + 1.5*a)
t.pendown()
t.color("blue","red")
t.begin_fill()
rectangular(a)
t.end_fill()

t.color("yellow","white")
t.begin_fill()
t.penup()
t.goto((-100+1.3*a),a)
t.pendown()
rectangular(a/3)
t.end_fill()

t.penup()
t.goto((-100+2*a),-100)
t.pendown()
t.color("blue","green")
t.begin_fill()
square(1.5*a)
t.end_fill()

t.penup()
t.goto((-100+2.4*a),-100)
t.pendown()
t.color("yellow","white")
rectangular(3/8*a)
t.fd(3/8*a)
rectangular(3/8*a)
```
