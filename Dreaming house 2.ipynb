```python
import turtle

t = turtle.Turtle()

turtle.pensize(5)
turtle.bgcolor("light blue")
t.penup()
t.goto(-250,-250)
t.pendown()

a = 50
b = 45
c = 30

#Vẽ nhà
t.fillcolor("blue")
t.begin_fill()
for i in range(4):
    t.fd(4*a)
    t.lt(b*2)
t.end_fill()

t.goto(-175,-250)
t.fillcolor("light green")
t.begin_fill()
for i in range(2):
    t.fd(a)
    t.lt(b*2)
    t.fd(a*2)
    t.lt(b*2)
t.end_fill()

t.penup()
t.goto(-250,-50)
t.pendown()
t.fillcolor("pink")
t.begin_fill()
t.fd(a*4)
t.lt(b*3)
t.fd(142)
t.lt(b*2)
t.fd(142)
t.end_fill()

t.penup()
t.goto(-50,-50)
t.pendown()
t.rt(b*2)
t.fillcolor("orange")
t.begin_fill()
for i in range(2):
    t.fd(142)
    t.rt(105)
    t.fd(145)
    t.rt(75)
t.end_fill()

t.rt(105)
t.fillcolor("yellow")
t.begin_fill()
t.fd(145)
t.rt(c*4)
t.fd(a*4)
t.rt(c*2)
t.fd(145)
t.end_fill()

t.penup()
t.goto(0,-75)
t.pendown()
t.lt(c*2)
t.fillcolor("red")
t.begin_fill()
t.fd(a)
t.lt(c*4)
t.fd(a)
t.lt(c*2)
t.fd(a)
t.end_fill()

# Vẽ cây
turtle.pensize(5)
turtle.bgcolor("light blue")
t.penup()
t.goto(200,-250)
t.pendown()
t.rt(b*2)

t.fillcolor("brown")
t.begin_fill()
for i in range(2):
    t.fd(a-10)
    t.lt(b*2)
    t.fd(a*2)
    t.lt(b*2)
t.end_fill()

t.penup()
t.goto(140,-150)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.fd(160)
t.lt(b*3)
t.fd(114)
t.lt(b*2)
t.fd(114)
t.end_fill()


t.penup()
t.goto(150,-68)
t.pendown()
t.lt(b*3)
t.fillcolor("green")
t.begin_fill()
t.fd(140)
t.lt(b*3)
t.fd(a*2)
t.lt(b*2)
t.fd(a*2)
t.end_fill()

t.penup()
t.goto(160,5)
t.pendown()
t.lt(b*3)
t.fillcolor("green")
t.begin_fill()
t.fd(c*4)
t.lt(b*3)
t.fd(85)
t.lt(b*2)
t.fd(85)
t.end_fill()

# Vẽ ông mặt trời
t.penup()
t.goto(100,150)
t.pendown()
t.lt(b*3)
t.fillcolor("yellow")
t.begin_fill()
t.circle(a)
t.end_fill()
t.rt(b*2)
t.fd(a)
t.rt(b*4)
t.fd(a)

t.penup()
t.fd(a*2)
t.pendown()
t.fd(a)
t.rt(b*4)
t.fd(a)

t.rt(86)
t.penup()
for i in range(40):
    t.fd(1)
    t.lt(1)
t.pendown()
t.rt(c*3)
t.fd(c)
t.rt(c*6)
t.fd(c)
t.penup()
t.fd(a*2)
t.pendown()
t.fd(c)
t.rt(c*6)
t.fd(c)

t.rt(86)
t.penup()
for i in range(40):
    t.fd(1)
    t.lt(1)
t.pendown()
t.rt(c*3)
t.fd(a)
t.rt(c*6)
t.fd(a)
t.penup()
t.fd(a*2)
t.pendown()
t.fd(a)
t.rt(c*6)
t.fd(a)

t.rt(86)
t.penup()
for i in range(40):
    t.fd(1)
    t.lt(1)
t.pendown()
t.rt(c*3)
t.fd(c)
t.rt(c*6)
t.fd(c)
t.penup()
t.fd(a*2)
t.pendown()
t.fd(c)
t.rt(c*6)
t.fd(c)
```
